data understanding / visualisation

```python
def data_info(x, y = 1):
    #print data info
    print(x.info())
    print(10*"-----")
    print(x.describe())
    print(10*"-----")
    print("sample:")
    print(x.sample(5))
    print(msno.matrix(x))
    if y == 0:
        print(10*"-----")
        print("timestamp start:" + x.timestamp.min())
        print("timestamp end:" + x.timestamp.max())
    print(10*"-----")
    print("count of missing values:")
    print(x.isnull().sum())
    print(10*"-----")
    print("missing values visualized:")
```

```python
#train_df
data_info(train_df, 0)
```

```python
#test_df
data_info(test_df, 0)
```

```python
#building_df
data_info(building_df)
```
